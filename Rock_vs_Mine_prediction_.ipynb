importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

data collection and data processing

```python
#gettin the data set to padas dataframe
sonar_data=pd.read_csv('/content/Copy of sonar data.csv',header=None)
```

```python
sonar_data.head()
```

```python
# number of rows and cols
sonar_data.shape
```

```python
sonar_data.describe() #statistical measure
```

```python
sonar_data[60].value_counts()
```

M--> MIne
R-->Rock


```python
sonar_data.groupby(60).mean()
```

```python
#separating data on label
x= sonar_data.drop(columns=60,axis=1)
y=sonar_data[60]
```

```python
print(x)
print(y)
```

Training and test data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y, random_state=1)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model Training

```python
model=LogisticRegression()
```

```python
#training logic regression model with training data
model.fit(x_train,y_train)
```

model evaluation

```python
#find accuracy on training data
x_train_prediction=model.predict(x_train)
training_data_accuracy =accuracy_score(x_train_prediction,y_train)
```

```python
print('Acuuracy of training data: ',training_data_accuracy )
```

```python
#find accuracy on test data
x_test_prediction=model.predict(x_test)
test_data_accuracy =accuracy_score(x_test_prediction,y_test)
```

```python
print('Acuuracy of test data: ',test_data_accuracy )
```

```python
import streamlit as st
import numpy as np
import joblib

# Load the model
model = joblib.load("rock_mine_model.pkl")

# App title
st.title("Rock vs. Mine Predictor")

# Input fields
input_data = []
for i in range(60):
    value = st.number_input(f"Feature {i+1}", min_value=0.0, max_value=1.0, value=0.0, step=0.01)
    input_data.append(value)

# Predict button
if st.button("Predict"):
    input_data_np = np.array(input_data).reshape(1, -1)
    prediction = model.predict(input_data_np)

    # Display result
    if prediction[0] == 'R':
        st.write("Object is a Rock")
    else:
        st.write("Object is a Mine")
```
