Importing the dependancies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
#data collection and analysis
#PIMA Diabetes Dataset

diabtes_dataset=pd.read_csv('/content/diabetes.csv')
```

```python
diabtes_dataset.groupby('Outcome').mean()
```

```python
#seprate data and label
x=diabtes_dataset.drop(columns='Outcome',axis=1)
y=diabtes_dataset['Outcome']
```

```python
print(x)
print(y)
```

```python
#Data Standardization
scaler=StandardScaler()
```

```python
scaler.fit(x)
```

```python
standadized_data=scaler.transform(x)
```

```python
print(standadized_data)
```

```python
x=standadized_data
y=diabtes_dataset['Outcome']
```

```python
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Training the model

```python
classifier=svm.SVC(kernel='linear')
```

```python
#training the support vector machine clssifier
classifier.fit(x_train,y_train)
```

model evaluation

```python
#accuracy score

x_train_prediction=classifier.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy score of training data:',training_data_accuracy)
```

```python
x_test_prediction=classifier.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy score of test data:',test_data_accuracy)
```

Making a predictive system

```python
input_data=(11,143,94,33,146,36.6,0.254,51)

#change into a numpy array

input_data_as_numpy_array=np.asarray(input_data)

#reshape the array as we are predicting
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)

#standardize the input data
std_data=scaler.transform(input_data_reshaped)

prediction=classifier.predict(std_data)
print(prediction)

if (prediction[0]==1):
  print('the person in diabetic')
else:
  print('the person is not diabetic')
```

```python
import pickle

# Save classifier
with open('classifier.pkl', 'wb') as file:
    pickle.dump(classifier, file)

# Save scaler
with open('scaler.pkl', 'wb') as file:
    pickle.dump(scaler, file)
```

```python
import numpy as np
import pandas as pd
import pickle
import ipywidgets as widgets
from IPython.display import display

# Load the model and scaler
with open('classifier.pkl', 'rb') as model_file:
    classifier = pickle.load(model_file)

with open('scaler.pkl', 'rb') as scaler_file:
    scaler = pickle.load(scaler_file)

# Define the input fields as widgets
pregnancies = widgets.FloatText(description="Pregnancies")
glucose = widgets.FloatText(description="Glucose")
blood_pressure = widgets.FloatText(description="Blood Pressure")
skin_thickness = widgets.FloatText(description="Skin Thickness")
insulin = widgets.FloatText(description="Insulin")
bmi = widgets.FloatText(description="BMI")
dpf = widgets.FloatText(description="Diabetes Pedigree Function")
age = widgets.FloatText(description="Age")
predict_button = widgets.Button(description="Predict")

# Define the output display area
output = widgets.Output()

# Define the feature names matching those used during training
feature_names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
                 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

# Define the prediction function
def predict_diabetes(b):
    with output:
        output.clear_output()

        # Collecting data from user inputs
        input_data = [
            pregnancies.value,
            glucose.value,
            blood_pressure.value,
            skin_thickness.value,
            insulin.value,
            bmi.value,
            dpf.value,
            age.value
        ]

        # Convert input data to a DataFrame with feature names
        input_data_df = pd.DataFrame([input_data], columns=feature_names)

        # Standardize the input data
        std_data = scaler.transform(input_data_df)

        # Prediction
        prediction = classifier.predict(std_data)

        # Display the result
        if prediction[0] == 1:
            print("The person is diabetic")
        else:
            print("The person is not diabetic")

# Bind the button click to the prediction function
predict_button.on_click(predict_diabetes)

# Display all widgets
display(pregnancies, glucose, blood_pressure, skin_thickness, insulin, bmi, dpf, age, predict_button, output)
```
